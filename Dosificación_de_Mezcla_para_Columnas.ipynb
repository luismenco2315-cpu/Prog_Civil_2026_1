#**Dosificación de Mezcla para Columnas**

**Planteamiento**:

Un ingeniero residente necesita calcular cuántos sacos de cemento se requieren para vaciar una columna de sección circular.

 Se sabe que para un concreto estándar de 210 kg/cm2 (3000 psi), se consumen 9.5 sacos de cemento por cada metro cúbico de mezcla.

**Requerimientos**:

    1. Solicitar la altura de la columna en metros.
    2. Solicitar el diámetro de la columna en metros.
    3. Calcular el volumen de la columna:
  $$v = \pi * r^2 * h$$.
    4. (Sugerencia: usar 3.1416 o importar math.pi).
    
 $$sacos = v * 9.5$$
   

Prueba del programa:

Para una columna de 3 metros de alto y 0.4 metros de diámetro (r=0.2):

    Volumen ≈ 0.377 m3.
    Sacos necesarios ≈ 3.58

````python
# ENTRADA:Solicitar la altura de la columna en metros.
# Solicitar el diametro de la columna en metros.
h = float(input("Ingrese la altura de la columna en metros: "))
d = float(input("Ingrese el diámetro de la columna en metros: "))

# PROCESO: Calcula el radio r, calcula el volumen de la columna, calcula sacos
r = d/2
v = 3.1416 * r**2 * h
sacos = v * 9.5

# SALIDA: Imprime el resultado
print(f"Volumen ≈ {v:.3f} m3.")
print(f"Sacos necesarios ≈ {sacos:.2f}")
````

```markdown
Aquí está la explicación de cada línea de código en el script:

*   `h = float(input("Ingrese la altura de la columna en metros: "))`:
    *   `input(...)`: Muestra el mensaje "Ingrese la altura de la columna en metros: " al usuario y espera que ingrese un valor.
    *   `float(...)`: Convierte la entrada del usuario (que siempre es una cadena) a un número de punto flotante (decimal).
    *   `h = ...`: Asigna este valor numérico a la variable `h`, que representa la altura de la columna.

*   `d = float(input("Ingrese el diámetro de la columna en metros: "))`:
    *   Similar a la línea anterior, pero solicita el diámetro de la columna y lo almacena como un número de punto flotante en la variable `d`.

*   `r = d/2`:
    *   Calcula el radio de la columna dividiendo el diámetro (`d`) entre 2.
    *   El resultado se almacena en la variable `r`.

*   `v = 3.1416 * r**2 * h`:
    *   Esta línea calcula el volumen de la columna circular utilizando la fórmula $$v = \pi * r^2 * h$$.
    *   `3.1416`: Es una aproximación del valor de Pi (π).
    *   `r**2`: Calcula el radio (`r`) elevado al cuadrado.
    *   `* h`: Multiplica el resultado por la altura (`h`).
    *   El volumen calculado se guarda en la variable `v`.

*   `sacos = v * 9.5`:
    *   Calcula la cantidad total de sacos de cemento necesarios.
    *   Multiplica el volumen de la columna (`v`) por la constante `9.5`, que representa la cantidad de sacos de cemento requeridos por metro cúbico de mezcla.
    *   El número final de sacos se almacena en la variable `sacos`.

*   `print(f"Volumen ≈ {v:.3f} m3.")`:
    *   Muestra el volumen calculado al usuario.
    *   `f"..."`: Es una cadena formateada (f-string) que permite insertar variables directamente en la cadena.
    *   `{v:.3f}`: Formatea el valor de la variable `v` para mostrarlo con 3 decimales.

*   `print(f"Sacos necesarios ≈ {sacos:.2f}")`:
    *   Muestra la cantidad de sacos de cemento necesarios al usuario.
    *   `{sacos:.2f}`: Formatea el valor de la variable `sacos` para mostrarlo con 2 decimales.
```
