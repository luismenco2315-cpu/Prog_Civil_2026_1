# 1. Problema sencillo de ingeniería

Supongamos un caso muy común en **ingeniería civil, agrícola o agroindustrial**:

**Calcular el volumen de agua almacenado en un tanque cilíndrico.**

Esto sirve por ejemplo para:

* Tanques de riego
* Reservorios de agua
* Tanques de almacenamiento en procesos industriales

La fórmula del volumen del cilindro es:

genui{"math_block_widget_always_prefetched": {"content": "V = π r^2 h"}}

Donde:

* **V** = volumen
* **r** = radio del tanque
* **h** = altura del tanque
* **π ≈ 3.1416**

---

# 2. Definir Entradas – Proceso – Salidas

### Entradas (Inputs)

Datos que el usuario debe proporcionar:

* radio del tanque (r)
* altura del tanque (h)

---

### Proceso

1. Leer el radio
2. Leer la altura
3. Calcular el volumen usando la fórmula
4. Guardar el resultado en una variable

---

### Salidas (Outputs)

* Mostrar el **volumen del tanque**

---

# 3. Algoritmo

**Algoritmo en lenguaje natural**

1. Inicio
2. Leer radio
3. Leer altura
4. Calcular volumen = π × radio² × altura
5. Mostrar volumen
6. Fin

---

# 4. Pseudocódigo

```
INICIO

leer radio
leer altura

volumen = π * radio^2 * altura

mostrar volumen

FIN
```

---

# 5. Código en Python (básico)

```python
# -----------------------------------------
# Cálculo del volumen de un tanque cilíndrico
# Aplicación básica de variables y operadores
# -----------------------------------------

import math   # biblioteca que contiene el valor de pi

# 1. Entradas
radio = float(input("Ingrese el radio del tanque (m): "))
altura = float(input("Ingrese la altura del tanque (m): "))

# 2. Proceso
volumen = math.pi * radio**2 * altura

# 3. Salida
print("El volumen del tanque es:", volumen, "m3")
```

### Un ejemplo aplicado a la Ingeniería

Imagina que necesitas calcular el **volumen de agua** necesario para llenar un tanque de almacenamiento cilíndrico en una finca.

Antes de mostrarte el código, pensemos en la lógica (algoritmo):

1. Definir el radio y la altura del tanque.
2. Aplicar la fórmula matemática: $V = \pi \cdot r^2 \cdot h$.
3. Mostrar el resultado con unidades claras.

Aquí tienes un ejemplo de cómo se vería esto siguiendo **buenas prácticas** (variables descriptivas y modularización):

import math

def calcular_volumen_tanque(radio_m, altura_m):
    """Calcula el volumen de un cilindro en metros cúbicos."""
    volumen = math.pi * (radio_m ** 2) * altura_m
    return volumen

# Datos de entrada
radio_tanque = 2.5  # metros
altura_tanque = 5.0 # metros

# Llamada a la función
resultado_v = calcular_volumen_tanque(radio_tanque, altura_tanque)

print(f"El volumen del tanque es: {resultado_v:.2f} m³")
