## **Historia, ciencia y programación: la “pluma de agua”**

> **Relato de partida**  
> **¿Por qué los costeños le decimos “pluma” a la llave del agua?**  
>  
> En el Caribe colombiano, abrir la “pluma” no es solo una acción cotidiana: es una expresión cargada de historia, técnica y cultura. Aunque algunos se sorprendan o incluso se burlen del término, lo cierto es que decir “pluma” en vez de “llave” no es un error, sino una herencia viva que conecta nuestro lenguaje con siglos de evolución hidráulica.  
>  
> Mucho antes de que existieran los grifos modernos, el agua se distribuía mediante sistemas regulados con precisión. En España, por ejemplo, se usaba la **“pluma de agua”** como unidad de medida para calcular el caudal que se entregaba a cada hogar. Aproximadamente, una pluma equivalía a **0.025 litros por segundo**. En algunos lugares, ese flujo se regulaba mediante orificios calibrados y dispositivos llamados **plumeros**, que permitían controlar la salida del agua.  
>  
> Durante la época colonial, muchas de estas prácticas técnicas cruzaron el Atlántico y llegaron a ciudades como Cartagena, Barranquilla y Santa Marta. Con el tiempo, el término “pluma” dejó de ser solamente técnico y pasó al habla cotidiana como sinónimo de la llave del agua.  
>  
> Así, una palabra del uso diario conserva una memoria histórica, tecnológica y cultural.

## 2. Situación problema

Un hogar recibe una cierta cantidad de **plumas de agua**.  
Se sabe que:

- **1 pluma = 0.025 litros por segundo**

Con esta información, el estudiante debe construir un programa que permita calcular:

1. el caudal total en litros por segundo;
2. el consumo diario en litros;
3. el consumo mensual en litros (suponiendo 30 días);
4. una clasificación del consumo:
   - **Bajo**: menos de 1000 litros por día
   - **Medio**: entre 1000 y 3000 litros por día
   - **Alto**: más de 3000 litros por día

Complete el siguiente bloque de código a partir del algoritmo anterior.

````python
# ACTIVIDAD: MODELAR LA "PLUMA DE AGUA" CON PYTHON
# ==========================================================

# 1. Solicitar al usuario el número de plumas
plumas = float(input("Ingrese el número de plumas: "))

# 2. Definir constantes del problema
# Caudal de una pluma en litros por segundo
CAUDAL_PLUMA = 0.025 # litros/segundo
# Segundos en un día
SEGUNDOS_DIA = 24 * 60 * 60 # segundos
# Días en un mes para el cálculo (asumiendo 30 días)
DIAS_MES = 30

# 3. Calcular el caudal total (litros por segundo)
caudal_total = plumas * CAUDAL_PLUMA

# 4. Calcular el consumo diario (litros)
consumo_diario = caudal_total * SEGUNDOS_DIA

# 5. Calcular el consumo mensual (litros)
consumo_mensual = consumo_diario * DIAS_MES

# 6. Clasificar el consumo
nivel_consumo = ""
if consumo_diario < 1000:
    nivel_consumo = "Bajo"
elif 1000 <= consumo_diario <= 3000:
    nivel_consumo = "Medio"
else:
    nivel_consumo = "Alto"

# 7. Mostrar resultados
print(f"\n--- Resultados del Consumo de Agua ---")
print(f"Número de plumas: {plumas}")
print(f"Caudal total: {caudal_total:.3f} litros por segundo")
print(f"Consumo diario estimado: {consumo_diario:.2f} litros")
print(f"Consumo mensual estimado (30 días): {consumo_mensual:.2f} litros")
print(f"Clasificación del consumo: {nivel_consumo}")
````

## Implementación de Extensiones Opcionales

````python
# Nuevas constantes para las extensiones
DIAS_SEMANA = 7
TARIFA_POR_LITRO = 0.005 # Ejemplo: 0.005 unidades monetarias por litro
LITROS_POR_METRO_CUBICO = 1000 # 1 metro cúbico = 1000 litros
````

````python
def calcular_consumo_hogar(num_plumas, caudal_pluma=0.025, segundos_dia=86400, dias_mes=30, dias_semana=7, tarifa_por_litro=0.005, litros_por_metro_cubico=1000):
    """
    Calcula el consumo de agua de un hogar y lo clasifica.

    Args:
        num_plumas (float): Número de plumas de agua del hogar.
        caudal_pluma (float): Caudal de una pluma en litros por segundo.
        segundos_dia (int): Segundos en un día.
        dias_mes (int): Días en un mes.
        dias_semana (int): Días en una semana.
        tarifa_por_litro (float): Costo por litro de agua.
        litros_por_metro_cubico (int): Litros en un metro cúbico.

    Returns:
        dict: Un diccionario con los resultados del cálculo y la clasificación.
    """
    caudal_total = num_plumas * caudal_pluma
    consumo_diario = caudal_total * segundos_dia
    consumo_semanal = consumo_diario * dias_semana
    consumo_mensual = consumo_diario * dias_mes

    consumo_diario_m3 = consumo_diario / litros_por_metro_cubico
    consumo_semanal_m3 = consumo_semanal / litros_por_metro_cubico
    consumo_mensual_m3 = consumo_mensual / litros_por_metro_cubico

    costo_mensual = consumo_mensual * tarifa_por_litro

    nivel_consumo = ""
    if consumo_diario < 1000:
        nivel_consumo = "Bajo"
    elif 1000 <= consumo_diario <= 3000:
        nivel_consumo = "Medio"
    else:
        nivel_consumo = "Alto"

    return {
        'Número de Plumas': num_plumas,
        'Caudal Total (L/s)': caudal_total,
        'Consumo Diario (L)': consumo_diario,
        'Consumo Diario (m³)': consumo_diario_m3,
        'Consumo Semanal (L)': consumo_semanal,
        'Consumo Semanal (m³)': consumo_semanal_m3,
        'Consumo Mensual (L)': consumo_mensual,
        'Consumo Mensual (m³)': consumo_mensual_m3,
        'Costo Mensual': costo_mensual,
        'Clasificación': nivel_consumo }
````

### Comparación del consumo de varios hogares

````python
import pandas as pd

# Definir una lista de hogares con diferentes números de plumas
hogares_plumas = [
    {'Hogar': 'Hogar A', 'Plumas': 1},
    {'Hogar': 'Hogar B', 'Plumas': 5},
    {'Hogar': 'Hogar C', 'Plumas': 10},
    {'Hogar': 'Hogar D', 'Plumas': 2},
    {'Hogar': 'Hogar E', 'Plumas': 7}]

# Calcular el consumo para cada hogar
resultados_hogares = []
for hogar in hogares_plumas:
    num_plumas = hogar['Plumas']
    resultados = calcular_consumo_hogar(num_plumas, CAUDAL_PLUMA, SEGUNDOS_DIA, DIAS_MES, DIAS_SEMANA, TARIFA_POR_LITRO, LITROS_POR_METRO_CUBICO)
    resultados['Hogar'] = hogar['Hogar']
    resultados_hogares.append(resultados)

# Convertir los resultados a un DataFrame de pandas para presentación tabular
df_comparacion = pd.DataFrame(resultados_hogares)

# Reordenar las columnas para mejor legibilidad
df_comparacion = df_comparacion[['Hogar', 'Número de Plumas', 'Caudal Total (L/s)',
                                   'Consumo Diario (L)', 'Consumo Diario (m³)',
                                   'Consumo Semanal (L)', 'Consumo Semanal (m³)',
                                   'Consumo Mensual (L)', 'Consumo Mensual (m³)',
                                   'Costo Mensual', 'Clasificación']]

# Mostrar los resultados
display(df_comparacion)
````

````python
# Nuevas constantes para las extensiones
DIAS_SEMANA = 7
TARIFA_POR_LITRO = 0.005 # Ejemplo: 0.005 unidades monetarias por litro
LITROS_POR_METRO_CUBICO = 1000 # 1 metro cúbico = 1000 litros

def calcular_consumo_hogar(num_plumas, caudal_pluma=0.025, segundos_dia=86400, dias_mes=30, dias_semana=7, tarifa_por_litro=0.005, litros_por_metro_cubico=1000):
    """
    Calcula el consumo de agua de un hogar y lo clasifica.

    Args:
        num_plumas (float): Número de plumas de agua del hogar.
        caudal_pluma (float): Caudal de una pluma en litros por segundo.
        segundos_dia (int): Segundos en un día.
        dias_mes (int): Días en un mes.
        dias_semana (int): Días en una semana.
        tarifa_por_litro (float): Costo por litro de agua.
        litros_por_metro_cubico (int): Litros en un metro cúbico.

    Returns:
        dict: Un diccionario con los resultados del cálculo y la clasificación.
    """
    # 3. Calcular el caudal total (litros por segundo)
    caudal_total = num_plumas * caudal_pluma

    # 4. Calcular el consumo diario (litros)
    consumo_diario = caudal_total * segundos_dia

    # Calcular consumo semanal (Litros y Metros Cúbicos)
    consumo_semanal = consumo_diario * dias_semana

    # 5. Calcular el consumo mensual (litros)
    consumo_mensual = consumo_diario * dias_mes

    # Convertir a metros cúbicos
    consumo_diario_m3 = consumo_diario / litros_por_metro_cubico
    consumo_semanal_m3 = consumo_semanal / litros_por_metro_cubico
    consumo_mensual_m3 = consumo_mensual / litros_por_metro_cubico

    # Estimar el costo del agua
    costo_mensual = consumo_mensual * tarifa_por_litro

    # 6. Clasificar el consumo
    nivel_consumo = ""
    if consumo_diario < 1000:
        nivel_consumo = "Bajo"
    elif 1000 <= consumo_diario <= 3000:
        nivel_consumo = "Medio"
    else:
        nivel_consumo = "Alto"

    return {
        'Número de Plumas': num_plumas,
        'Caudal Total (L/s)': caudal_total,
        'Consumo Diario (L)': consumo_diario,
        'Consumo Diario (m³)': consumo_diario_m3,
        'Consumo Semanal (L)': consumo_semanal,
        'Consumo Semanal (m³)': consumo_semanal_m3,
        'Consumo Mensual (L)': consumo_mensual,
        'Consumo Mensual (m³)': consumo_mensual_m3,
        'Costo Mensual': costo_mensual,
        'Clasificación': nivel_consumo
    }

import pandas as pd

# Definir una lista de hogares con diferentes números de plumas
hogares_plumas = [
    {'Hogar': 'Hogar A', 'Plumas': 1},
    {'Hogar': 'Hogar B', 'Plumas': 5},
    {'Hogar': 'Hogar C', 'Plumas': 10},
    {'Hogar': 'Hogar D', 'Plumas': 2},
    {'Hogar': 'Hogar E', 'Plumas': 7}
]

# Calcular el consumo para cada hogar
resultados_hogares = []
for hogar in hogares_plumas:
    num_plumas = hogar['Plumas']
    resultados = calcular_consumo_hogar(num_plumas, CAUDAL_PLUMA, SEGUNDOS_DIA, DIAS_MES, DIAS_SEMANA, TARIFA_POR_LITRO, LITROS_POR_METRO_CUBICO)
    resultados['Hogar'] = hogar['Hogar']
    resultados_hogares.append(resultados)

# Convertir los resultados a un DataFrame de pandas para presentación tabular
df_comparacion = pd.DataFrame(resultados_hogares)

# Reordenar las columnas para mejor legibilidad
df_comparacion = df_comparacion[['Hogar', 'Número de Plumas', 'Caudal Total (L/s)',
                                   'Consumo Diario (L)', 'Consumo Diario (m³)',
                                   'Consumo Semanal (L)', 'Consumo Semanal (m³)',
                                   'Consumo Mensual (L)', 'Consumo Mensual (m³)',
                                   'Costo Mensual', 'Clasificación']]

# Mostrar los resultados
display(df_comparacion)
````
