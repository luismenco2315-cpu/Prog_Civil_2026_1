## **Teoría del Confort Térmico en Ganado Vacuno**

El confort térmico en ganado vacuno se refiere a un estado en el que los animales no experimentan estrés por calor o frío. Mantener el ganado dentro de su zona de confort térmico es crucial para optimizar la producción (leche, carne) y asegurar su bienestar.

### **Importancia del Confort Térmico**

Los bovinos son sensibles a las variaciones de temperatura y humedad. Cuando el ambiente excede sus límites de confort, pueden sufrir estrés térmico, lo que lleva a:

*   **Reducción en la producción:** Disminución en la ingesta de alimento, menor producción de leche, crecimiento más lento.
*   **Problemas de salud:** Mayor susceptibilidad a enfermedades.
*   **Problemas reproductivos:** Afectación de la fertilidad.

### **Índice de Temperatura y Humedad (ITH)**

El ITH es una de las herramientas más comunes para evaluar el riesgo de estrés térmico en el ganado. Combina la temperatura del aire y la humedad relativa para proporcionar un valor único que refleja la sensación térmica del animal.

La fórmula utilizada en este análisis es:

$$ITH = 0.72 * (T_{bs} + T_{bh}) + 40.6$$

Donde:

*   **Tbs:** Temperatura de bulbo seco (°C). Es la temperatura del aire medida con un termómetro estándar.
*   **Tbh:** Temperatura de bulbo húmedo (°C). Es la temperatura que indica la capacidad de enfriamiento del aire por evaporación, influenciada por la humedad.

### **Interpretación del ITH (según National Weather Service)**

Los valores de ITH se correlacionan con diferentes niveles de estrés térmico:

1.  **ITH ≤ 70:** **Ambiente no estresante.** El ganado se encuentra en su zona de confort térmico.
2.  **70 < ITH ≤ 78:** **Situación de alerta por estrés térmico.** Los animales pueden empezar a experimentar estrés leve y se recomienda monitoreo y posibles medidas preventivas.
3.  **78 < ITH ≤ 83:** **Peligro por estrés térmico.** El estrés térmico es significativo y se requieren acciones para mitigar sus efectos.
4.  **ITH > 83:** **Situación crítica.** Los animales están en alto riesgo de sufrir consecuencias graves debido al calor, y se necesitan intervenciones urgentes para proteger su salud y producción.

# **Métodos y cálculos**

+ Índice de temperatura y humedad (producción intensiva)

$$ITH = 0.72 * (T_{bs} + T_{bs}) + 40.6$$

Donde:

1. Tbs: Temperatura bulbo seco (°C)
2. Tbh: Temperatura de bulbo húmedo (°C)

Para la correlación de los datos de la especie bovina en la producción de leche se recomienda
utilizar los límites propuestos por National Weather Service:

1. Sí $ITH <= 70$ ambiente no estresante
2. Sí $ITH > 70 \space and \space $ITH <= 78$ situación de alerta por estrés térmico

3. Sí $ITH > 78 \space and \space <= 83$ peligro por estrés térmico


4. Sí $ITH > 83$ situación crítica


````python
# ENTRADA: Pedir tbs (Temperatura de  bulbo seco) y  tbh (Temperatura de bulbo humedo)
tbs = float(input("Ingrese la temperatura de bulbo seco, Grados centigrados: "))
tbh = float(input("Ingrese la temperatura de bulbo húmedo, grados centigrados : "))

# PROCESO: Calculo de ith indice de temperatura y humedad)
ith = 0.72 * (tbs + tbh) + 40.6

# SALIDA: Imprimir resultado, de la evaluacion del ith y el confort térmico
if ith <= 70:
    print(f"El ambiente no es estresante")
elif ith > 70 and ith <= 78:
    print(f"La situacion es de alerta por estrés térmico")
else:
    print(f"La situacion es de critico por estrés térmico")

````
